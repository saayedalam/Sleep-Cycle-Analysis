## ‚úÖ A/B Testing Workflow (Condensed)

1. **Define Objective**

2. **Load Data**  
   - Import dataset.  
   - Parse datetime columns if needed.

3. **Clean Data**  
   - Fix missing values, data types, and units (e.g., seconds ‚Üí hours).  
   - Drop irrelevant or unreliable columns.

   - Visualize distributions.  
   - Detect outliers and assess group balance.

5. **Define Hypotheses**  
   - H‚ÇÄ: No difference between groups  
   - H‚ÇÅ: There is a difference

8. **Run Statistical Test**  
   - Use t-test, Welch‚Äôs t-test, or Mann-Whitney U test depending on distribution.

9. **Interpret Results**
